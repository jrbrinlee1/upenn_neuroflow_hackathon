# Code to build video space

Grabbed all text related with video (desctiption, transcripts, tags, etc.) and build 20,000 features using the most common words. Joined this with categorical information built by machine learning model.

```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer
```

```python
video_data = pd.read_excel("processed_video_data_for_bag_of_words.xlsx")
```

```python
video_data = video_data.set_index('video_id')
```

```python
videofeature = pd.read_csv("video_features.csv")
```

```python
cntVtr = CountVectorizer(ngram_range = (1,1), max_features = 20000)
```

```python
X = cntVtr.fit_transform(list(video_data['as string']))
```

```python
video_cnt_vtr = pd.DataFrame(X.toarray())
```

```python
video_cnt_vtr['video_id'] = video_data.index
```

```python
video_final = pd.merge(video_cnt_vtr, videofeature, on = "video_id")
```

```python
video_final = video_final.set_index('video_id')
```

```python
video_final.describe()
```

```python
video_final.to_csv("video_space.csv")
```
